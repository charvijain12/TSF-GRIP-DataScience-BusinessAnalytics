```python
import json
import folium
import requests
import mimetypes
import http.client
import pandas as pd
from folium.plugins import HeatMap
from pandas.io.json import json_normalize
```

### Access Data via API

```python
conn=http.client.HTTPSConnection("api.covid19api.com")
```

```python
payload=''
headers={}
conn.request("GET","/summary",payload,headers)
res=conn.getresponse()
data=res.read().decode('UTF-8')
```

### Convert Data to JSON

```python
covid1=json.loads(data)
pd.json_normalize(covid1['Countries'],sep=',')
```

```python
df=pd.DataFrame(covid1['Countries'])
df
```

```python
covid2=df.drop(columns=['CountryCode','Slug','Date','Premium'],axis=1)
covid2
```

### Generating BaseMap

```python
m=folium.Map(tiles="Stamen Terrain",min_zoom=1.5)
m
```

### Obtain Geodata

```python
geo='world-countries.json'
```

### Generate Choropleth Map Layer 

```python
m.choropleth(
    geo_data=geo,
    name='Covid-19',
    data=covid2,
    columns=['Country','TotalConfirmed'],
    key_on='feature.properties.name',
    fill_color='OrRd',
    legend_name='Total Confirmed Covid Cases'
)

m
```

### Generate Circular Markers

```python
covid2.update(covid2['TotalConfirmed'].map('Total Confirmed:{}'.format))
covid2.update(covid2['TotalRecovered'].map('Total Recovered:{}'.format))
coordinates = pd.read_csv('countries.csv', encoding='ISO-8859-1')
coordinates
```

```python
coordinates=coordinates.rename(columns={'name':'Country','country':'country-code'})
coordinates
```

```python
covid_final=pd.merge(covid2,coordinates)
covid_final.head()
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(covid_final.latitude, covid_final.longitude):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
```

```python
# add pop-up text to each marker on the map
latitudes = list(covid_final.latitude)
longitudes = list(covid_final.longitude)
labels = list(covid_final.Country)
tc = list(covid_final.TotalConfirmed)
tr = list(covid_final.TotalRecovered)

for lat, lng, label, Total, Recover in zip(latitudes, longitudes, labels, tc, tr):
    folium.Marker([lat, lng], popup=label + Total + Recover).add_to(m)    
    
    
```

```python
# add incidents to map
m.add_child(incidents)
```
